```python
import numpy as np
import matplotlib.pyplot as plt
```

# 9.Decision Trees

```python
#Entropy: Im = -Sum(p * log p)
p = np.linspace(0, 1, 200)[1:-1]
I = - p * np.log2(p)
I2 = - p * np.log2(p) - (1-p) * np.log2(1-p)

plt.figure(1, figsize=(12, 6))

plt.subplot(121)
plt.plot(p, I)
plt.xlim(0, 1)
plt.xlabel('p')
plt.ylabel('- p * log p')

plt.subplot(122)
plt.plot(1-p, I)
plt.xlim(0, 1)
plt.xlabel('q')
plt.ylabel('- q * log q')


plt.figure(2)
plt.plot(p, I2)
plt.xlim(0, 1)
plt.xlabel('p')
plt.ylabel('- p * log p - (1-p) * log(1-p)')
plt.vlines(x=0.5, ymin=0, ymax=1, linestyle='dotted')

plt.show()
```

# 10. Linear Discrimination

```python
def sig(x):
  return 1 / (1 + np.exp(-x))

def line(a, x, b):
  return a * x + b

x = np.linspace(0, 5, 200)

plt.ylim(-1, 3)
plt.plot(x, line(-30, x, 80))
plt.plot(x, sig(line(-30, x, 80)))
```

```python
def pl(x, y):
  return x + 2*y + [3, 3]

x = [1, 5]
y = [2, 9]
plt.plot(x, y, pl(x, y))
```
